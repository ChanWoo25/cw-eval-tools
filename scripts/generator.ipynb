```python
def process(img_fn: Path):
    img = cv2.imread(str(img_fn), cv2.IMREAD_ANYCOLOR)
    img = cv2.resize(img, (3200, 3200))
    print("[Image]")
    print("- path: %s" % img_fn.name)
    print("- type: ", type(img))
    print("- shape: ", img.shape)
    print("- dtype: ", img.dtype)
    w = img.shape[0]
    h = img.shape[1]
    
    pallete = np.ones((img_width, img_height, 3), dtype=np.uint8) * 255
    print("[Pallete]")
    print("- type: ", type(pallete))
    print("- shape: ", pallete.shape)
    print("- dtype: ", pallete.dtype)
    
    cx = img_width  // 2
    cy = img_height // 2
    
    pallete[cx-w//2:cx+w//2, cy-h//2:cy+h//2] = img
    return pallete
    
```
